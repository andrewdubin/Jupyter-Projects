```python
#data dependencies
import pandas as pd
import numpy as np
from pandas import DataFrame, Series, read_html
#web scraping
import requests
from bs4 import BeautifulSoup
```

```python
df = pd.read_csv('responsecodelist.csv')
df
```

```python
#excluding all entries without boobpedia pages
df = df[df['Response Codes']=='<Response [200]>']
df
```

```python
test_url = 'http://boobpedia.com/boobs/Mia_Malkova'
result = requests.get(test_url)
c = result.content
soup = BeautifulSoup(c)
soup
```

```python
# Go to the section of interest
info = soup.find('table',{'class':'infobox'})
info
```

```python
# Set up empty data list
data = []

# Set rows as first indexed object in tables with a row
rows = info.findAll('tr')

# now grab every HTML cell in every row
for tr in rows:
    cols = tr.findAll('td')
    # Check to see if text is in the row
    for td in cols:
        text = td.find(text=True) 
        data.append(text)

data
```

```python
#lists to fill
ethnicity_list = []
nationality_list = []
measurements_list = []
brasize_list = []
boob_list = []


for i in df['Boobpedia URL']:
    data = []
    #get to the site
    result = requests.get(i)
    c = result.content
    soup = BeautifulSoup(c)
    # Go to the section of interest
    info = soup.find('table',{'class':'infobox'})
    # Set rows as first indexed object in tables with a row
    rows = info.findAll('tr')
    # now grab every HTML cell in every row
    for tr in rows:
        cols = tr.findAll('td')
        # Check to see if text is in the row
        for td in cols:
            text = td.find(text=True) 
            data.append(text)
    ethnicity_list.append(data[9])
    nationality_list.append(data[11])
    measurements_list.append(data[14])
    brasize_list.append(data[16])
    boob_list.append(data[18])

df['Ethnicity'] = ethnicity_list
df['Nationality'] = nationality_list
df['Measurements'] = measurements_list
df['Bra Size'] = brasize_list
df['Boobs'] = boob_list
df
```
