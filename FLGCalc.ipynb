```python
# Import libraries
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
# url we're going to scrape their names from
url = 'https://www.fleshlight.com/collections/fleshlight-girls'
# Connection to the URL
response = requests.get(url)
# Parse HTML and save to BeautifulSoup object
soup = BeautifulSoup(response.text, "html.parser")
```

```python
girls = soup.find('section',{'class':'products'})
girls = girls.find_all('h3')
girls = list(girls)
```

```python
#helper function to remove the heading tags from the list of girls
def remove_h3tags(the_list):
    new_list = []
    for i in the_list:
        i = str(i)
        i = i[4:-5]
        new_list.append(i)
    return new_list
girls = remove_h3tags(girls)
```

Time to put their names and sleeve prices into a DataFrame. Sleeve prices on their own are 59.95 a piece, 79.95 with the container. All of these girls have pussy sleeves, but not all of them have butt sleeves. By default, we'll set all the values to the sleeve price, and then zero out the entries for stars without butt sleeves.

```python
df = pd.DataFrame({'Name':girls,'Lady Sleeve':59.95,'Butt Sleeve':59.95})
df
```

Now to calculate the cost of buying all of the sleeves without containers. After adding at least ten items to your cart on their website, a 15% discount is applied to every individual item in the cart. Use this url to add items to your cart https://www.fleshlight.com/collections/orgy. The math below in the Cost columns reflects that discount

```python
df['Cost'] = round(( (df['Lady Sleeve']*.85) + (df['Butt Sleeve']*.85) ),2)
subtotal = df['Cost'].sum()
#account for taxes
total = round(subtotal *1.06,2)
print('Subtotal ' + str(subtotal) + '\n Total ' + str(total))
```
